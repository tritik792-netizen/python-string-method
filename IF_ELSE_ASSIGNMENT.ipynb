```python
units = int(input("Enter number of units: "))
bill = 0

# First 100 units: no charge
if units > 100:
    # Next 100 units (101–200)
    if units > 200:
        bill = 100 * 5   # for units 101–200
        bill = bill + (units - 200) * 10  # above 200 units
    else:
        bill = (units - 100) * 5

print("Total bill amount is Rs", bill)
```

```python
salary = float(input("Enter your salary: "))
years = int(input("Enter years of service: "))

bonus = 0

if years > 10:
    bonus = salary * 0.10

if years >= 6 and years <= 10:
    bonus = salary * 0.08

if years < 6:
    bonus = salary * 0.05

print("Bonus amount is:", bonus)
```

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

# Assume a is second largest initially
second = a

if (a > b and a < c) or (a < b and a > c):
    second = a

if (b > a and b < c) or (b < a and b > c):
    second = b

if (c > a and c < b) or (c < a and c > b):
    second = c

print("Second largest number is:", second)
```

```python
eng = int(input("Enter marks in English: "))
math = int(input("Enter marks in Math: "))
sci = int(input("Enter marks in Science: "))
sst = int(input("Enter marks in Social Studies: "))

# Science Stream
if eng > 80 and math > 80 and sci > 80 and sst > 80:
    print("Science Stream")

# Commerce Stream
if eng > 80 and math > 50 and sci > 50:
    print("Commerce Stream")

# Humanities Stream
if eng > 80 and sst > 80:
    print("Humanities")
```
